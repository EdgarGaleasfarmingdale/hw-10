```python
# Array Observation will store the list of observations of covid cases
# Array Prediction will store the list of predictions of covid cases

Observation = [1, 0, 1, 1, 0, 1, 0, 0, 0, 1]
Prediction = [1, 1, 1, 0, 0, 1, 0, 0, 1, 1]

# TP will store the count of true positive cases
# TN will store the count of true negative cases
# FP will store the count of false positive cases
# FN will store the count of false negative cases

TP = 0
TN = 0
FP = 0
FN = 0

# for-loop is used to iterate through the observation and prediction array
# each observation and corresponding prediction is compared within the loop
# to identify TP, TN, FP and FN
# at the end of the loop, total number of TP, TN, FP and FN is available

for i in range(0,len(Observation)):
    if Observation[i] == 1 and Prediction[i] == 1:
       TP = TP + 1
    if Observation[i] == 0 and Prediction[i] == 0:
       TN = TN + 1
    if Observation[i] == 0 and Prediction[i] == 1:
       FP = FP + 1
    if Observation[i] == 1 and Prediction[i] == 0:
       FN = FN + 1

print("Observation : ", Observation)
print("Prediction  : ", Prediction)

# print is used to display TP, TN, FP and FN values

print("TP = ", TP)
print("TN = ", TN)
print("FP = ", FP)
print("FN = ", FN)

# Accuracy = (True Positive + True Negative) / (True Positive + True 
# Negative + False Positive + False Negative)

# Precision = True Positive / (True Positive + False Positive)

# Recall = True Positive / (True Positive + False Negative)

#F1-score = 2 * ((Precision * Recall) / (Precision + Recall))

Accuracy = (TP + TN) / (TP + TN + FP + FN)
Precision = TP / (TP + FP)
Recall = TP / (TP + FN)
F1 = 2 * (Precision * Recall) / (Precision + Recall)

# print is used with appropriate formatting to print upto 2 digits after 
# the decimal places

print("Accuracy : ", "{:.2f}".format(Accuracy*100))
print("Precision : ", "{:.2f}".format(Precision*100))
print("Recall : ", "{:.2f}".format(Recall*100))
print("F1 Score : ", "{:.2f}".format(F1*100))
```

Accuracy, precision, recall and F1 scores are the general metrics used to measure the performance of a classification model.

Let us assume an example dataset of 10 COVID-19 patients and the prediction of each patient of the dataset.

Observation = [1, 0, 1, 1, 0, 1, 0, 0, 0, 1]

Prediction = [1, 1, 1, 0, 0, 1, 0, 0, 1, 1]

Observation = 1 Means patient is COVID-19 Positive

Observation = 0 Means patient is COVID-19 Negative

Prediction = 1 Means the patient is predicted to be COVID-19 Positive

Prediction = 0 Means the patient is predicted to be COVID-19 Negative

Observations can be classified into 4 categories: True Positive (TP), False Positive (FP), True Negative (TN), False Negative (FN)

True Positive (TP) = Patient is COVID-19 positive and the predicted result is also positive (Observation = Prediction = 1)

False Positive (FP) = Patient is not COVID-19 positive but the predicted result is positive (Observation = 0 and Prediction = 1)

True Negative (TN) = Patient is not COVID-19 positive and the predicted result is also not positive (Observation = Prediction = 0)

False Negative (FN) = Patient is COVID-19 positive but the predicted result is negaative (Observation = 1 and Prediction = 0)

Accuracy: Accuracy is a measure that determines how many observations are predicted correctly. It can be computed by dividing the sum of (TP + TN) by the sum of (TP + TN + FP + FN)

Accuracy = (True Positive + True Negative) / (True Positive + True Negative + False Positive + False Negative)

Precision: Precision is a measure that determines how many observations that are marked as positive are actually positive. It can be computed by dividing the TP by the sum of TP + FP

Precision = True Positive / (True Positive + False Positive)

Recall: Recall is also known as sensitivity. It is a measure that determines how many positive observations were actually classified as positive. It can be computed by dividing the TP by the sum of TP + FN

Recall = True Positive / (True Positive + False Negative)

F1-Score: The F1-score is a combination of precision and recall as a single metric and is computed by calculating the harmonic mean of precision and recall.

F1-score = 2 * ((Precision * Recall) / (Precision + Recall))
